```python
# Grouping Numerical Data

df_num = df[['Age','DailyRate','DistanceFromHome','Education','EmployeeCount','EmployeeNumber','RelationshipSatisfaction',
            'StandardHours','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance',
             'YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
df_num
```

```python
# Grouping Categorical Data

df_categorical = df[['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole',
                     'MaritalStatus','Over18','OverTime']]
```

```python
df_categorical
```

```python
# For simpler caliculation converting character of attrition column into binaru form
num_val = {'Yes':1, 'No':0}
df_categorical['Attrition'] = df_categorical["Attrition"].apply(lambda i:num_val[i])
```

```python
df_categorical.head()
```

```python
df_categorical = pd.get_dummies(df_categorical)
df_categorical.head()
```

```python
# merging numerical & categorical daya by concatinating

df_data = pd.concat([df_num, df_categorical], axis=1)
```

```python
df_data.head()
```

```python
from sklearn.model_selection import train_test_split

target = df_data['Attrition']
predictors = df_data.drop("Attrition", axis=1)
X_train,X_test,Y_train,Y_test = train_test_split(predictors,target,test_size=0.20,random_state=0)
```

```python
# Caliculating accuracy score with algorithms

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

Lreg=LogisticRegression()
Lreg.fit(X_train,Y_train)
Y_pred_Lreg = Lreg.predict(X_test)
Y_pred_Lreg.shape
```

```python
score_Lreg = round(accuracy_score(Y_pred_Lreg,Y_test)*100,2)
print("The Attrition Accuracy Through Logistic Regression is :"+str(score_Lreg)+"%")
```
